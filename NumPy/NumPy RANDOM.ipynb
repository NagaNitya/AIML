# Learning NumPy Random Functions

Random numbers are those which cannot be predicted or put in a sequence. Computers use algorithms to generate randome numbers; these are called *pseudo-random* numbers.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
from numpy import random # module in NumPy to generate random numbers
```

Random number may be generated using the `random` module in *NumPy*. The methods to create random numbers are:
+ `random.rand()` for float
+ `random.randint()` for int

These may be applied to generate an array of random numbers as well, by passing the array size as parameters. Besides these methods, we can use the `random.choice()` to select a random element of a given array.

```python
print(random.randint(1, 100)) # generate a random integer between 1 and 100
print(random.rand()) # generate a random float between 0 and 1
print(random.randint(50, size=(3, 4)))
print(random.rand(4))
print(random.choice([2, 3, 4, 5, 7, 8, 1]))
```

### Data Distribution

A *data distribution* is a list of all possible values, as well as how often each value occurs.

A *random distribution* is a set of random numbers that follow a *probability density function*. A probability density function is a function that describes a continuous probability.

We can generate random numbers on defined probabilities by using the `random.choice()` method. Here there are 3 main parameters:
+ The array of number to choose randomly from.
+ `p` ie. the probability that each number will be included.
+ `size` ie. the number of values to generate.

```python
# generate 100 values from the given numbers [3, 5, 7, 9]
# the probability of each value occurring in the final distribution is given by the 'p' array. eg. 3 and 10% chance of occurring.
x=random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))
print(x)
```

Notice that the number 9 did not appear at all in the distribution. This is because the probability of it as given in the `p` array is 0.

Also note that the sum of probabilities must equal 1. ie. `0.1 + 0.3 + 0.6 + 0.0 = 1.0`

```python
z = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))
print(z)
```

### Random Permutations

A *permutation* refers to an arrangement of elements. The `random` module offers 2 methods for generatiing permuatations of numbers:
+ `random.shuffle()` which shuffles the elements of the array in place.
+ `random.permutation()` which generates a random permutation of the given array.


```python
arr=np.array([1, 2, 3, 4, 5])
random.shuffle(arr) # shuffle the array in place
print(arr)

arr2=random.permutation(arr) # return a new array with the elements shuffled
print(arr2)
```

### Normal Distribution

The *Normal Distribution* is one of the most important random distributions. It is characterised by a bell shaped curve, and is used to portray the probability distributions of events such as IQ scores, heartbeat rates, etc.

the `random.normal()` method is used to crerate a normal distribution. This method has 3 parameters: 
+ `loc`: the mean or the peak of the curve
+ `scale`: standard deviation, or how flat the curve is
+ `size`: shape of the returned array

```python
# generate a 2x3 matrix with random numbers from a normal distribution with mean 0 and standard deviation 1
norm=random.normal(loc=0, scale=1, size=(2, 3)) 
print(norm)
```

```python
# plot the above matrix as a distplot using seaborn
sns.distplot(norm, hist=False) # hist=False will not plot the histogram
plt.show()
```

### Binomial Distribution

It is a *discrete distribution* that describes the outcome of binary scenarios eg. tossing of a coin.

Use the `random.binomial()` method to create a binomial distribution. This method has 3 parameters:
+ `n`: number of trials.
+ `p`: probability of occurence of each trial.
+ `size`: the shape of the returned array.

```python
# generate a 2x5 matrix in which each element/event goes through 10 trials. 
# The probability of success of each trial is 0.5
binom = random.binomial(n=10, p=0.5, size=(2, 5))
print(binom)
```

Here each element undergoes 10 trials (`n=10`). with each trial having a specified probability of 0.5 (`p=0.5`), it counts the sumber of successful trials and returns these as a 2x5 matrix (`size=(2, 5)`).

```python
sns.distplot(binom, hist=False)
plt.show()
```

POISSON DISTRIBUTION

This a type of discrete distribution that estimates how many times an event can occur in a specified time interval. Use the `random.poisson()` method to create a poisson distribution. This method has 2 parameters:
+ `lam`: rate or number of known occurences.
+ `size`: shape of the returned array.

```python
# create a poisson distribution with known lambda value of 2 and generate 10 random numbers
pois = random.poisson(lam=2, size=10)
print(pois)
```

The poisson distibution showcases the randomness around an average. Given that an event occurs 2 times in a time interval, the poisson distribution showcases the randomness associated with this number. ie. the event may occur once or thrice in certain instances as well.

```python
sns.distplot(pois, hist=False)
plt.show()
```

UNIFORM DISTRIBUTION

The uniform distribution specifies that all numbers within a specific range have equal probability of occurence. eg. a coin toss, in which the numbers 1-6 each have an equal probability of occurrence.

We use the `random.uniform()` method which has 3 parameters:
+ `low`: lower bound with default=0.0. (inclusive)
+ `high`: upper bound with default 1.0. (exclusive)
+ `size`: shape of the returned array.

```python
# generate 10 numbers in the range 0 to 5 with equal probabilities.
uni=random.uniform(high=5, size=10) 
print(uni)
```

the output depicts 10 numbers selected with equal probability, from the range [0, 5).

```python
sns.distplot(uni, hist=False)
plt.show()
```

LOGISTIC DISTRIBUTION

The logistic distribution is used to desribe growth. It finds use in the logistic regression algorithm in ML, neural networks, etc.

Use the `random.logistic()` method to create a logistic distribution. This has 3 parameters:
+ `loc`: where the mean or peak is. Default value is 0.
+ `scale`: standard deviation. Default is 1.
+ `size`: shape of the returned array.

```python
log = random.logistic(loc=1, scale=2, size=(2, 3))
print(log)
```

```python
sns.distplot(log, hist=False)
plt.show()
```

### Multinomial Distribution

A multinomial distribution is a generalisation of the binomial distribution, which gives the outcomes for multinomial scenarios, eg. rolling of a dice.

Use the `random.multinomial()` function to create a multinomial distiribution, with the following 3 parameters:
+ `n`: number of possible outcomes.
+ `pvals`: the probability value of each outcome.
+ `size`: shape of the returned array.

```python
mult=random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6]) # generate 10 samples of 6 trials, imitating a dice throw
print(mult)
```

```python
sns.distplot(mult, hist=False)
plt.show()
```

### Exponential Distribution

It is used for describing the time till the next event.

the exponential distribution is made using the `random.exponential()` method. This has 2 parameters:
+ `scale`: the inverse of rate. 
+ `size`: shape of the returned array.


```python
exp=random.exponential(scale=2, size=10) # generate 10 random numbers from an exponential distribution with lambda=2
print(exp)
```

```python
sns.distplot(exp, hist=False)
plt.show()
```

### Chi Square Distribution 

This is commonly used to verify the hypothesis. The method used is `random.chisquare()` and there are 2 parameters:
+ `df`: degrees of freedom.
+ `size`: shape of the returned array.

```python
chi=random.chisquare(df=2, size=10) # generate 10 random numbers from a chi-square distribution with 2 degrees of freedom
print(chi)
```

```python
sns.distplot(chi, hist=False)
plt.show()
```
